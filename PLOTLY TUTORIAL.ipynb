```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = pd.read_csv("datasets\Plotly\multipleChoiceResponses.csv",encoding='ISO-8859-1');
data.head()
```

```python
# é¥¼çŠ¶å›¾
gender =  data['GenderSelect'].value_counts()
Label = gender.values
Size = gender.index
plt.figure(figsize = (12,12));
plt.axes(aspect = 1)
plt.pie(Label,labels = Size,explode = [0,0.3,0.3,0.2],autopct = "%.3f");
plt.legend(loc="best")
plt.show()
```

```python

# é¥¼çŠ¶å›¾
gender =  data['FirstTrainingSelect'].value_counts()
colors = ['aqua', 'lightgrey', 'lightgreen', '#D0F9B1', 'khaki', 'grey']
Label = gender.values
Size = gender.index
plt.figure(figsize = (12,12));
plt.axes(aspect = 1)
plt.pie(Label,labels = Size,autopct = "%.3f",colors = colors);
plt.legend(loc="best")
plt.title('First Training Platform')
plt.show()
```

```python
hardware = data['HardwarePersonalProjectsSelect'].str.split(',');
hardware_set = []
for i in hardware.dropna():
    hardware_set.extend(i);
hware = pd.Series(hardware_set).value_counts()[:6] 
label = hware.index
size = hware.values
ex = [0.01,0.01,0.01,0.01,0.3,0.01]
fig,ax = plt.subplots(figsize = (10,10));
ax.pie(size,autopct="%.2f",shadow=True,explode = ex,labels = label);
ax.legend(loc =2)
plt.title("Hardware Requirements")
plt.show()
```
